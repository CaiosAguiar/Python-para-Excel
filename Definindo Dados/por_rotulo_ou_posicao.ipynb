## Substituição de valores

```python
rainfall = pd.DataFrame(data={"City 1": [300.1, 100.2],
                              "City 2": [400.1, 500.3],
                              "City 3": [1000.5, 1100.6]})
# copiando original
rain_search = rainfall.copy()
# substituindo valor caso o valor seja true(verdadeiro)
rain_search[rain_search < 400] = 0
```

```python
# para substituir determinado valor em todo o DataFrame ou em colunas, use replace()
df2.replace("USA", "U.S.")
 
# para atuar em colunas especificas, dessa o replace funciona por meio de chave e valor 
# replace({"coluna": {"valor_antigo": "novo_valor"}})
df.replace({"country": {"USA": "U.S."}})
```
