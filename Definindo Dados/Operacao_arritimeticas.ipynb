# Operadores Aritiméticos

```python
import pandas as pd
import numpy as np
```

```python
rainfall = pd.DataFrame(data={"City 1": [300.1, 100.2],
                              "City 2": [400.1, 500.3],
                              "City 3": [1000.5, 1100.6]})
# copiando original
rain_search = rainfall.copy()

more_rain = pd.DataFrame(data=[[100, 200], [300, 400]],
                         index=[1, 2],
                         columns=["City 1", "City 4"])
```

### Operador de adição usando add (+)

```python
# use o fill=value para substituir os NaN 
rain_search.add(more_rain, fill_value=0)
```

### Operador de subtração usando sub (-)

```python
less_rain = more_rain.copy()
rain_search.sub(less_rain, fill_value=0)
```

### Operador de multiplicação usando mul (*)

```python
df1 = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [10, 20, 30, 40]
})

df2 = pd.DataFrame({
    "A": [5, np.nan, 7, 8],
    "B": [1, 2, np.nan, 4]
})
```

```python
df1.mul(df2, fill_value=1)
```

### Operação de divisão usando div (/)

```python
df1.div(df2, fill_value=1)
```

### Operação de potenciação usando pow (**)

```python
df1.pow(df2)
```
