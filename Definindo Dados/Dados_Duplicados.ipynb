Dados Duplicados

```python
import pandas as pd
df = pd.read_excel(r"C:\Users\caios\OneDrive\Desktop\Python para Excel\Cápitulo 5\Planilhas\course_participants.xlsx")
```

```python
# para se livrar de linnhas duplicadas, use drop_duplicates
# podemos fonecer um subconjunto de colunas
df.drop_duplicates(["country", "continent"])
```

```python
# saber o resultado de duplicatas no índice
df.index.is_unique

# is_unique é um atributo que retorna verdadeiro se todos os elementos são unicos
# ou falso se existem duplicados
df["country"].is_unique

# retorna um array com todos os valores únicos presentes na Série
df["country"].unique()
# array(['Italy', 'USA', 'Germany'], dtype=object)
```

### duplicated
#### retorna uma Série booleana com as linhas que estão duplicadas
#### por padrão a primeira linha/ocorrência do elemento é dado como (False)
#### use keep="False" para retorna True para todas as linhas, incluindo sua primeira ocorrência 

```python
# ele marca apenas a duplicatas como True, sem a primeira ocorrência
df["country"].duplicated()
```

```python
# para obter todas as linhas em que o elemento está duplicado use keep="False"
df.loc[df["country"].duplicated(keep="False"), :]
```
