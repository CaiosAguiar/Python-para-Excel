```python
data = [["Mark", 55, "Italy", 4.5, "Europe"],
        ["John", 33, "USA", 6.7, "America"],
        ["Tim", 41, "USA", 3.9, "America"],
        ["Jenny", 12, "Germany", 9.0, "Europe"]]

df = pd.DataFrame(data=data,
                  columns=["name", "age", "country",
                           "score", "continent"],
                  index=[1001, 1000, 1002, 1003])
df

print(df)

# Várias Colunas / DataFrame 
# df.loc[:, ["country", "age"]]
```

# SELECIONANDO POR RÓTULO
### reordenar colunas com .loc[: ["colunas reodenadas..."]]
### : → seleciona todas as linhas. Porém os dois pontos (:) por último significam selecionar por colunas
### [ ... ] → define a ordem das colunas explicitamente.

```python
colunas_reordenadas = df.loc[:,[ "continent", "country", "name", "age", "score"]]
print(colunas_reordenadas)
```

```python
# Valor Único / Tipo de dado retornado: escalar
tabela_pais = (df.loc[1000, "country"])
```

```python
# Intervalo de colunas / DataFrame
df.loc[:, "name":"country"]
```

```python
# Uma linha (1d) / Série
df.loc[1000, :]
```

```python
# Uma linha (2d) / DataFrame
df.loc[[1000], :]
```

```python
# Várias linhas / DataFrame
df.loc[[1003, 1000], :]
```

# FATIAMENTO DE RÓTULOS

```python
# Usar escalares para a seleção de linha e coluna retorna um escalar
df.loc[1000, "name"]
```

```python
#Usar um escalar na seleção de linha ou coluna retorna uma série
df.loc[[1000, 1002], "age"]
```

```python
# Selecionar várias linhas e colunas retorna um DataFrame
df.loc[:1001, ["name", "country"]]
```

# SELECIONANDO POR INDEXAÇÃO BOOLEANA

## and == & --- or == | --- not == ~  

```python
tf = (df_planilha["age"] > 40) & (df_planilha["country"] == "USA") # tf é uma Série com apenas True/False
df_planilha.loc[tf, :]
```

```python
# Se quiser filtar o índice, você pode se referir a ele como df.index
df.loc[df.index > 1001, :]
```

## .isin()
### isin() serve para verificar se os valores de uma coluna ou Série estão contidos em uma lista (ou em outro objeto iterável).

```python

df.loc[df["country"].isin(["Italy", "Germany"]), :]
```

## DataFrames com Base em Números
### caso algum número não corresponda com o critério retorna NaN(False) ou True caso corresponda

```python
rainfall = pd.DataFrame(data={"City 1": [300.1, 100.2],
                              "City 2": [400.1, 500.3],
                              "City 3": [1000.5, 1100.6]})
```

```python
rainfall > 400
```

```python
rainfall[rainfall < 400]
```

# SELECIONANDO COM UM MULTIINDEX

```python
# Um MultiIndex precisa ser ordenado
df_multi = df.reset_index().set_index(["continent", "country"])
df_multi = df_multi.sort_index()
df_multi
```

```python
# Selecionado por Tupla (não esqueça dos parênteses )
df_multi.loc[("Europe", "Italy"), :]
```

```python
# Para redefinir uma parte de um MultiIndex forneça o nível como argumento
# Zero é a primeira coluna da esquerda
df_multi.reset_index(level=0)
```
