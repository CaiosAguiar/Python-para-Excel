# Aplicando Funções

#### Diferença entre os métodos

##### .map() → só funciona em Series (coluna única).

##### .apply() → funciona em linhas ou colunas inteiras (axis=0 colunas, axis=1 linhas).

##### .applymap() → aplica função elemento a elemento em todo o DataFrame.

##### obs: applymap já está depreciado use .map() ao invés de .applymap()

```python
import pandas as pd
```

```python
littles = pd.DataFrame(
    data=[[3, 5, 3, 5],
          [3, 6, 2, 13],
          [14, 10, 3, 1]],
          index=[1, 2, 3],
          columns=["Mês 1", "Mês 2", "Mês 3", "Mês 4"]
)
```

```python
# usando .map() com lambda
# multiplicando lambda
df_multi = littles.map(lambda x: x * 10)
df_multi
```

```python
# usando lambda(lambda x: valor if (condição)) com if ternário
df_flag = littles.map(lambda x: "Alto" if x > 5 else "Baixo")
df_flag
```

### Comparação com .apply() por linha ou coluna

```python
# soma de cada linha (axis=1)
linha_soma = littles.apply(lambda row: row.sum(), axis=1)
linha_soma
```

```python
# Média de cada coluna (axis=0)
coluna_media = littles.apply(lambda col: col.mean(), axis=0)
coluna_media
```

# COPY() recapitulando o tema
#### No pandas, .copy() serve para criar uma cópia independente de um DataFrame ou Series.

#### Sem .copy(), muitas vezes o que você tem é apenas uma referência (um "espelho") do objeto original.

#### Com .copy(), você garante que as alterações não afetam o DataFrame original.
