# Concatenando
### Por padrão, concat junta os DataFrames nas linhas e alinha as colunas automaticamente

```python
import pandas as pd

df = pd.read_excel(r"C:\Users\caios\OneDrive\Desktop\Python para Excel\Cápitulo 5\Planilhas\course_participants.xlsx")
```

```python
data=[[15, "Brazil", 4.1, "Bruno"],
      [44, "France", 6.1, "Becky"]]
more_users = pd.DataFrame(data=data,
                          columns=["age", "country", "score", "name"],
                          index=[1000, 1011])
more_users
```

```python
# concat([DataFrames que serão concatenado], axis=0)
pd.concat([df, more_users], axis=0)
```
